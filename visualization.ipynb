# load data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris=sns.load_dataset("iris")
```

```python
iris.head(1)
```

# Line plot

```python
plt.figure(figsize=(6,4))
plt.plot(iris['sepal_length'],color='red',label='sepal length')
plt.plot(iris['sepal_width'],color='green',label = 'sepal width')
plt.title('Iris Sepal Length')
plt.xlabel('Indexes')
plt.ylabel('Length Width')
plt.legend(loc='best')
plt.savefig("iris")
plt.show()
```

# Bar plot

```python
tip=sns.load_dataset("tips")
```
